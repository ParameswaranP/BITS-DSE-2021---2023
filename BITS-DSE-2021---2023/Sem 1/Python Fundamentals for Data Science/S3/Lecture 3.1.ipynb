```python
mark_list = [10,45,23,56,78]

total_marks = 0 
student_count = len(mark_list)

index = 0 
while index < student_count:
    #do the summation 
    total_marks = total_marks + mark_list[index]
    index = index + 1

print("Total marks is ", total_marks)

avg = total_marks / student_count

print("Average is ", avg)
```

<b> break & continue statement <b>

```python
i = 0
while i<10:
    i = i + 1
    if(i==5):
        break
    print("i = ", i)
print("I am outside ...")
```

```python
i = 0
while i<10:
    i = i + 1
    if(i==5):
        continue
    print("i = ", i)
print("I am outside ...")
```

```python
list1 = [1,2,3,4,5]

number = int(input("Enter the number to search: "))

index = 0

while index <len(list1):
    if(number == list1[index]):
        print("Number present !!")
        break
    #index = index + 1

    
print("I am outside the loop")
```
